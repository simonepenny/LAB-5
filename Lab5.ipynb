```python
dt = pd.read_csv('summer-travel-gps-full.csv', index_col='date', parse_dates=True)
dt.head()
```

```python
dt.index.is_unique
```

```python
dt = dt[~dt.index.duplicated(keep='first')]
dt
```

```python
dt.index.is_unique
```

```python
# Q18: create a weekday and a weekend dataframe
dt = dt.reset_index()
dt['weekday'] = dt['date'].dt.dayofweek
dt
```

```python
weekday_df = dt[(dt['weekday'] == 0) |  (dt['weekday'] == 1) |  (dt['weekday'] == 2 ) |  (dt['weekday'] == 3) |  (dt['weekday'] == 4)]
weekday_df
```

```python
weekend_df = dt[(dt['weekday'] == 5) |  (dt['weekday'] == 6) ]
weekend_df
```
